# Model Validation Techniques

```python
RND1 = smf.ols('RND~Admin+MKS',data=df1).fit().rsquared  
RND = 1/(1-RND1) 
RND
```

```python
Admin1 = smf.ols('Admin~RND+MKS',data=df1).fit().rsquared  
Admin = 1/(1-Admin1) 
Admin
```

```python
MKS1 = smf.ols('MKS~RND+Admin',data=df1).fit().rsquared  
MKS = 1/(1-MKS1) 
MKS
```

# Residual Q-Q Plot

```python
import statsmodels.api as sm
sns.set_style(style='darkgrid')
qqplot=sm.qqplot(model.resid,line='q') 
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-10000))
```

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "RND", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "Admin", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "MKS", fig=fig)
plt.show()
```

# Model Deletion Diagnostics

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
sns.set_style(style='darkgrid')
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(df1)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

# High Influence Plots

```python
fig,ax=plt.subplots(figsize=(15,10))
fig=influence_plot(model,ax = ax)
```
